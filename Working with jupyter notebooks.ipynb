   ### First lets open the training data using  `read.csv()`

```r
train = read.csv("../train.csv", header = T) 
head(train) ##head, by default prints the first five lines of a dataframe
```

To add a line click the `+` button on the tooltip 


Next lets get number in the 10th row and 3rd column of our variable `train` using the `[]` notation 

```r
train[10,3]
```

Lets get a bunch of things now in a single line: 

1) the first 10 rows and the first 3 columns 

2) rows 50 to 60 and the 4th and 5th columns 

3) and the Age column by it self

```r
train[1:10,1:3]
train[50:60,4:5]
train$Age
```

Notice how the output for each line printed out sequentially one after the other? 

Thats because in Jupyter notebooks each line is actually a block of code and when you click run, you run the whole block sequentially. 

Now lets remove all the rows in `train` with missing values using `complete.cases()`

```r
train.no.na = train[complete.cases(train),] ##remember that leaving the part after the comma blank returns all columns 
train.no.na$Age
```

Notice how when we printed out the Age column, there are no `<NA>` values anymore

Finally lets make the violin plots like last time using `GGPLOT`: 

```r
library(ggplot2) #load the ggplot2 library
p = ggplot(train.no.na, aes(x = factor(Survived), y = Age, fill = factor(Survived)))
p + geom_violin()
```
